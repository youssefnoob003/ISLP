**1. It was mentioned in this chapter that a cubic regression spline with one knot at $\xi$ can be obtained using a basis of the form $x, x^2, x^3, (x − \xi)^3_+$, where $(x − \xi)^3_+ = (x − \xi)^3$ if $x > \xi$ and equals 0 otherwise.**
**We will now show that a function of the form**
$$
f(x) = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \beta_4(x − \xi)^3_+
$$
**is indeed a cubic regression spline, regardless of the values of $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$.**

**(a) Find a cubic polynomial**
$$
f_1(x) = a_1 + b_1x + c_1x^2 + d_1x^3
$$
**such that $f(x) = f_1(x)$ for all $x \le \xi$. Express $a_1, b_1, c_1, d_1$ in terms of $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$.**

For $x \le \xi$, $\beta_4(x − \xi)^3_+ = 0 \Longrightarrow f(x) = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3$.
We can conclude that $a_1 = \beta_0, b_1 = \beta_1, c_1 = \beta_2, d_1 = \beta_3$

**(b) Find a cubic polynomial**
$$
f_2(x) = a_2 + b_2x + c_2x^2 + d_2x^3
$$
**such that $f(x) = f_2(x)$ for all $x \gt \xi$. Express $a_2, b_2, c_2, d_2$ in terms of $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$. We have now established that f(x) is a piecewise polynomial.**

For $x \gt \xi$, $\beta_4(x − \xi)^3_+ = \beta_4(x − \xi)^3 = \beta_4x^3 - 3\beta_4x^2\xi + 3\beta_4x\xi^2 - \beta_4\xi^3$
$$
\Longrightarrow f(x) = (\beta_0 - \beta_4\xi^3) + (\beta_1 + 3\beta_4\xi^2)x + (\beta_2 - 3\beta_4\xi)x^2 + (\beta_3 + \beta_4)x^3
$$

We can conclude that $a_2 = \beta_0 - \beta_4\xi^3, b_2 = \beta_1 + 3\beta_4\xi^2, c_2 = \beta_2 - 3\beta_4\xi, d_2 = \beta_3 + \beta_4$

**(c) Show that $f_1(\xi) = f_2(\xi)$. That is, $f(x)$ is continuous at $\xi$.**
$$
f_1(\xi) = a_1 + b_1\xi + c_1\xi^2 + d_1\xi^3
$$
$$
= \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3
$$
And:
$$
f_2(\xi) = (\beta_0 - \beta_4\xi^3) + (\beta_1 + 3\beta_4\xi^2)\xi + (\beta_2 - 3\beta_4\xi)\xi^2 + (\beta_3 + \beta_4)\xi^3
$$
$$
= \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3 + 3\beta_4\xi^3 - 3\beta_4\xi^3 - \beta_4\xi^3 + \beta_4\xi^3
$$
$$
= \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3 = f_1(\xi)
$$

We then conclude that $f(x)$ is continuous at $\xi$.

**(d) Show that $f'_1(\xi) = f'_2(\xi)$. That is, $f′(x)$ is continuous at $\xi$**

$$
f'_1(\xi) = b_1 + 2c_1\xi + 3d_1\xi^2
$$
$$
= \beta_1 + 2\beta_2\xi + 3\beta_3\xi^2
$$
And:
$$
f'_2(\xi) = (\beta_1 + 3\beta_4\xi^2) + 2(\beta_2 - 3\beta_4\xi)\xi + 3(\beta_3 + \beta_4)\xi^2
$$
$$
= \beta_1 + 2\beta_2\xi + 3\beta_3x^2 + 3\beta_4\xi^2 - 6 \beta_4\xi^2 + 3\beta_4\xi^2
$$
$$
= \beta_1 + 2\beta_2\xi + 3\beta_3\xi^2 = f'_1(\xi)
$$

We then conclude that $f'(x)$ is continuous at $\xi$.

**(e) Show that $f''_1(\xi) = f''_2(\xi)$. That is, $f''(x)$ is continuous at $\xi$**

$$
f''_1(\xi) = 2c_1 + 6d_1\xi
$$
$$
= 2\beta_2 + 6\beta_3\xi
$$
And:
$$
f''_2(\xi) = 2(\beta_2 - 3\beta_4\xi) + 6(\beta_3 + \beta_4)\xi
$$
$$
= 2\beta_2 + 6\beta_3\xi - 6\beta_4\xi + 6\beta_4\xi
$$
$$
= 2\beta_2 + 6\beta_3\xi
$$

We then conclude that $f''(x)$ is continuous at $\xi$.

**Therefore, f (x) is indeed a cubic spline.**

**2. Suppose that a curve $\hat{g}$ is computed to smoothly fit a set of n points using the following formula:**
$$
\hat{g} = arg \space \underset{g}{min} \left(\sum^n_{i=1}(y_i - g(x_i))^2 + \lambda \int \left[g^{(m)}(x)\right]^2 dx \right)
$$

where $g^{(m)}$ represents the mth derivative of g (and $g^{(0)} = g$). Provide
example sketches of $\hat{g}$ in each of the following scenarios.

**(a) $\lambda = \infty, m = 0$** In this case, g would be the null function, $g = 0$. Since $\lambda = \infty$, the function will only aim to minimize $\int \left[g^{(m)}(x)\right]^2 dx$, which reaches its minimum for g = 0 since we're computing the area under the curve.

**(b) $\lambda = \infty, m = 1$** Same as a, this will mainly aim to minimize the second part of the function, so, g must be a constant so that its dervative is null. the constant should be chosen to also minimize the first part, which is the mean of all the observations.

**(c) $\lambda = \infty, m = 2$** Same as a, this will mainly aim to minimize the second part of the function, so, g must be linear so that its second dervative is null. This effectively translates to a linear regression problem. since, after stating that g should be linear, we also minimize the first part.

**(d) $\lambda = \infty, m = 3$** Same as c, except this time, we can pick a second degree polynomial for more flexibility, which would translate to a normal linear regression.

**(d) $\lambda = 0, m = 3$** with $\lambda = 0$, the aim would be to minimize the first part only, which would result in a really wiggly function that would probably overfit the data since it doesnt have any constraints.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ISLP.models import poly, ModelSpec as MS 
import statsmodels.api as sm

# synthetic data
rng = np.random.default_rng(1)
n = 15
x = np.linspace(0, 10, n)
y = 2 + 0.5 * x + np.sin(x) + rng.normal(scale=1.0, size=n)

df = pd.DataFrame({'x': x, 'y': y})
x_grid = np.linspace(0, 10, 300)
df_grid = pd.DataFrame({'x': x_grid})

plt.figure(figsize=(14, 10))

# (a) lambda = inf, m = 0 -> g = 0
plt.subplot(231)
plt.scatter(x, y, c="black", label="Data")
plt.plot(x_grid, np.zeros_like(x_grid), "r", label="g=0")
plt.title("(a) λ=∞, m=0 (Zero function)")
plt.legend()

# (b) lambda = inf, m = 1 -> constant function (mean of y)
plt.subplot(232)
plt.scatter(x, y, c="black")
plt.plot(x_grid, np.full_like(x_grid, np.mean(y)), "r", label="Mean(y)")
plt.title("(b) λ=∞, m=1 (Constant)")
plt.legend()

# (c) lambda = inf, m = 2 -> linear regression
plt.subplot(233)
X = MS(['x']).fit_transform(df)
X_grid = MS(['x']).fit_transform(df_grid)
results = sm.OLS(y, X).fit()
y_lin = results.predict(X_grid)
plt.scatter(x, y, c="black")
plt.plot(x_grid, y_lin, "r", label="Linear fit")
plt.title("(c) λ=∞, m=2 (Linear regression)")
plt.legend()

# (d) lambda = inf, m = 3 -> quadratic regression
plt.subplot(234)
spec_quad = MS([poly('x', degree=2)])
X = spec_quad.fit_transform(df)  # fit on training data
df_grid = pd.DataFrame({'x': x_grid})
X_grid = spec_quad.transform(df_grid)  # just transform on new data

results = sm.OLS(y, X).fit()
y_quad = results.predict(X_grid)

plt.scatter(x, y, c="black")
plt.plot(x_grid, y_quad, "r", label="Quadratic fit")
plt.title("(d) λ=∞, m=3 (Quadratic regression)")
plt.legend()

# (e) lambda = 0, m = 3 -> interpolating spline (wiggly)
plt.subplot(235)
plt.scatter(x, y, c="black")
plt.plot(x, y, "r", label="Spline interpolant")
plt.title("(e) λ=0, m=3 (Overfit/interpolant)")
plt.legend()

plt.tight_layout()
plt.show()
```

**3. Suppose we fit a curve with basis functions $b_1(X) = X, b_2(X) = (X − 1)^2I(X \ge 1)$. (Note that $I(X \ge 1)$ equals 1 for $X \ge 1$ and 0 otherwise.) We fit the linear regression model:**
$$
Y = \beta_0 + \beta_1b_1(X) + \beta_2b_2(X) + \epsilon
$$
**and obtain coefficient estimates $\hat{\beta}_0 = 1, \hat{\beta}_1 = 1, \hat{\beta}_2 = −2$.**
**Sketch the estimated curve between $X = −2$ and $X = 2$. Note the intercepts, slopes, and other relevant information.**

For this problem, our estimated curve will take on a different form for each interval.

- $ -2 \le X \lt 1$: $b_1(X) = X$ and $b_2(X) = 0 \Longrightarrow \hat{Y} = 1 + X$ 
- $ 1 \le X \le 2$: $b_1(X) = X$ and $b_2(X) = (X - 1)^2 \Longrightarrow \hat{Y} = 1 + X - 2(X - 1)^2$ 

```python
def y_hat(x):
    if -2 <= x < 1:
        return 1 + x
    elif 1 <= x <= 2:
        return -2*x**2 + 5*x - 1
    else:
        return np.nan

X = np.linspace(-2, 2, 400)
Y = [y_hat(x) for x in X]

plt.figure(figsize=(8,5))
plt.plot(X, Y, color='blue', label=r'$\hat{Y}(X)$')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)

points = [-2, 0, 1, 1.25, 2]
for p in points:
    plt.scatter(p, y_hat(p), color='red')
    plt.text(p, y_hat(p)+0.2, f"({p:.2f},{y_hat(p):.2f})", ha='center')

plt.xlabel("X")
plt.ylabel(r"$\hat{Y}$")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

**4. Suppose we fit a curve with basis functions $b_1(X) = I(0 \le X \le 2) − (X − 1)I(1 \le X \le 2), b_2(X) = (X − 3)I(3 \le X \le 4) + I(4 \lt X \le 5).$. We fit the linear regression model:**
$$
Y = \beta_0 + \beta_1b_1(X) + \beta_2b_2(X) + \epsilon
$$
**and obtain coefficient estimates $\hat{\beta}_0 = 1, \hat{\beta}_1 = 1, \hat{\beta}_2 = 3$.**
**Sketch the estimated curve between $X = −2$ and $X = 6$. Note the intercepts, slopes, and other relevant information.**

For this problem, our estimated curve will take on a different form for each interval.

- $ -2 \le X \lt 0$: $b_1(X) = 0$ and $b_2(X) = 0 \Longrightarrow \hat{Y} = 1$ 
- $ 0 \le X \lt 1$: $b_1(X) = 1$ and $b_2(X) = 0 \Longrightarrow \hat{Y} = 1 + 1 = 2$ 
- $ 1 \le X \le 2$: $b_1(X) = 1 - (X - 1) = 2 - X$ and $b_2(X) = 0 \Longrightarrow \hat{Y} = 1 + 2 - X = 3 - X$
- $ 2 \lt X \lt 3$: $b_1(X) = 0$ and $b_2(X) = 0 \Longrightarrow \hat{Y} = 1$ 
- $ 3 \le X \le 4$: $b_1(X) = 0 and b_2(X) = X - 3 \Longrightarrow \hat{Y} = 1 + 3(X - 3) = 3X - 8$
- $ 4 \lt X \le 5$: $b_1(X) = 0 and b_2(X) = 1 \Longrightarrow \hat{Y} = 1 + 3 = 4$ 
- $ 5 \lt X \le 6$: $b_1(X) = 0 and b_2(X) = 0 \Longrightarrow \hat{Y} = 1$ 

```python
def y_hat(x):
    if -2 <= x < 0:
        return 1
    elif 0 <= x < 1:
        return 2
    elif 1 <= x <= 2:
        return 3 - x
    elif 2 < x < 3:
        return 1
    elif 3 <= x <= 4:
        return 3*x - 8
    elif 4 < x <= 5:
        return 4
    elif 5 < x <= 6:
        return 1
    else:
        return np.nan

X = np.linspace(-2, 6, 800)
Y = [y_hat(x) for x in X]

plt.figure(figsize=(10,6))
plt.plot(X, Y, label=r'$\hat{Y}(X)$', color='blue')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)

points = [-2,0,1,2,3,4,5,6]
for p in points:
    plt.scatter(p, y_hat(p), color='red')
    plt.text(p, y_hat(p)+0.2, f"({p},{y_hat(p)})", ha='center')

plt.xlabel("X")
plt.ylabel(r"$\hat{Y}$")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

$$
$$
\hat{g}_2 = arg \space \underset{g}{min} \left(\sum^n_{i=1}(y_i - g(x_i))^2 + \lambda \int \left[g^{(4)}(x)\right]^2 dx \right)
$$
**(a) As $\lambda \rightarrow \infty$, will $\hat{g}_1$ or $\hat{g}_2$ have the smaller training RSS?**

with $\lambda \rightarrow \infty$, the aim would be to minimize the second part, and this would be ideally done with $\lambda \int \left[g^{(m)}(x)\right]^2 dx = 0$. for $g_1$ this would be null up till a second degree polynominal, as opposed to $g_2$, which would stay null till a third degree polynominal. This would naturally allow $g_2$ to take up more complex shapes and be more flexible. Thus, $g_2$ would have a smaller training RSS, since it's more flexible.

**(b) As $\lambda \rightarrow \infty$, will $\hat{g}_1$ or $\hat{g}_2$ have the smaller test RSS?**

We stated that $g_2$ will be more flexible, yet since we don't know the nature of the problem, we can't say for sure which will have the lower test RSS.

**(c) For $\lambda = 0$, will $\hat{g}_1$ or $\hat{g}_2$ have the smaller training and test RSS?**

In this case, both functions would be minimizing the same target function, so both will give the same model, with the exact same test and training RSS.
