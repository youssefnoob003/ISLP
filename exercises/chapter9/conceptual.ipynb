**1. This problem involves hyperplanes in two dimensions.**

**(a) Sketch the hyperplane $1 + 3X_1 − X_2 = 0$. Indicate the set of points for which $1 + 3X_1 − X_2 \gt 0$, as well as the set of points for which $1 + 3X_1 − X_2 \lt 0$.**

**(b) On the same plot, sketch the hyperplane $−2 + X_1 + 2X_2 = 0$. Indicate the set of points for which $−2 + X_1 + 2X_2 \gt 0$, as well as the set of points for which $−2 + X_1 + 2X_2 \lt 0$.**

```python
import numpy as np
import matplotlib.pyplot as plt

x1_1 = np.linspace(-2, 2, 100)
x2_1 = 1 + 3*x1_1

pointsx1_1 = np.random.uniform(-2, 2, size=100)
pointsx2_1 = np.random.uniform(-4, 7, size=100)

lessx1_1, lessx2_1 = pointsx1_1[1 + 3*pointsx1_1 - pointsx2_1 < 0], pointsx2_1[1 + 3*pointsx1_1 - pointsx2_1 < 0]
morex1_1, morex2_1 = pointsx1_1[1 + 3*pointsx1_1 - pointsx2_1 > 0], pointsx2_1[1 + 3*pointsx1_1 - pointsx2_1 > 0]

plt.figure(figsize=(15, 9))

plt.plot(x1_1, x2_1, label=r"$x_2 = 1 + 3x_1$", color="blue")
plt.scatter(lessx1_1, lessx2_1, c="r", label=r"$1 + 3x_1 - x_2 < 0$")
plt.scatter(morex1_1, morex2_1, c="g", label=r"$1 + 3x_1 - x_2 > 0$")

x1_2 = np.linspace(-2, 2, 100)
x2_2 = 1 - 0.5 * x1_2

pointsx1_2 = np.random.uniform(-2, 2, size=100)
pointsx2_2 = np.random.uniform(-4, 7, size=100)

lessx1_2, lessx2_2 = pointsx1_2[-2 + pointsx1_2 + 2*pointsx2_2 < 0], pointsx2_2[-2 + pointsx1_2 + 2*pointsx2_2 < 0]
morex1_2, morex2_2 = pointsx1_2[-2 + pointsx1_2 + 2*pointsx2_2 > 0], pointsx2_2[-2 + pointsx1_2 + 2*pointsx2_2 > 0]

plt.plot(x1_2, x2_2, label=r"$x_2 = 1 - \frac{1}{2} X_1$", color="purple")
plt.scatter(lessx1_2, lessx2_2, c="y", label=r"$−2 + X_1 + 2X_2 < 0$")
plt.scatter(morex1_2, morex2_2, c="k", label=r"$−2 + X_1 + 2X_2 > 0$")


plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.title(r"Points Classified by Lines $1 + 3x_1 - x_2 = 0$ and $−2 + X_1 + 2X_2 = 0$")
plt.legend()
plt.grid(True)
plt.show()
```

**2. We have seen that in $p = 2$ dimensions, a linear decision boundary takes the form $\beta0 +\beta_1 X_1 +\beta_2 X_2 = 0$. We now investigate a non-linear decision boundary.**

**(a) Sketch the curve**
$$
(1 + X_1)² + (2 - X_2)² = 4
$$

**(b) On your sketch, indicate the set of points for which**

$$
(1 + X_1)² + (2 - X_2)² \gt 4
$$

**as well as the set of points for which**

$$
(1 + X_1)² + (2 - X_2)² \le 4
$$

```python
pointsx1 = np.random.uniform(-5, 3, size=300)
pointsx2 = np.random.uniform(-2, 6, size=300)

inside_mask = (1 + pointsx1)**2 + (2 - pointsx2)**2 <= 4
outside_mask = ~inside_mask

insidex1, insidex2 = pointsx1[inside_mask], pointsx2[inside_mask]
outsidex1, outsidex2 = pointsx1[outside_mask], pointsx2[outside_mask]

theta = np.linspace(0, 2*np.pi, 300)
circle_x = -1 + 2 * np.cos(theta)
circle_y =  2 + 2 * np.sin(theta)

plt.figure(figsize=(12, 8))

plt.plot(circle_x, circle_y, label=r"$(1+x_1)^2 + (2-x_2)^2 = 4$", color="purple")

plt.scatter(insidex1, insidex2, c="red", label=r"$(1+x_1)^2 + (2-x_2)^2 <= 4$")
plt.scatter(outsidex1, outsidex2, c="blue", label=r"$(1+x_1)^2 + (2-x_2)^2 > 4$")

plt.scatter(-1, 1, c='k', linewidths=3, label="(-1, 1)")
plt.scatter(2, 2, c='g', linewidths=3, label="(2, 2)")
plt.scatter(3, 8, c='y', linewidths=3, label="(3, 8)")

plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.title(r"Points Classified by $(1+x_1)^2 + (2-x_2)^2 = 4$")
plt.legend()
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
```

**(c) Suppose that a classifier assigns an observation to the blue class if $(1 + X_1)^2 + (2 − X_2)^2 > 4$, and to the red class otherwise. To what class is the observation (0, 0) classified? (−1, 1)? (2, 2)? (3, 8)?**

- $(-1, 1): (1 + (-1))^2 + (2 − 1)^2 = 1 < 4 \Longrightarrow$ **Red class**
- $(2, 2): (1 + 2)^2 + (2 − 2)^2 = 9 > 4 \Longrightarrow$ **Blue class**
- $(3, 8): (1 + 3)^2 + (2 − 8)^2 = 52 > 4 \Longrightarrow$ **Blue class**

**(d) Argue that while the decision boundary in (c) is not linear in terms of $X_1$ and $X_2$, it is linear in terms of $X_1$, $X_1^2$, $X_2$, and $X_2^2$.**

$$
(1 + X_1)^2 + (2 - X_2)^2 = 4.
$$

If we expand this expression, we get:

$$
(1 + X_1)^2 + (2 - X_2)^2 = 1 + 2X_1 + X_1^2 + 4 - 4X_2 + X_2^2 = X_1^2 + X_2^2 + 2X_1 - 4X_2 + 5 = 4,
$$

which simplifies to

$$
X_1^2 + X_2^2 + 2X_1 - 4X_2 + 1 = 0.
$$

Clearly, this is **not linear** in the original variables $X_1$ and $X_2$ because it contains quadratic terms $X_1^2$ and $X_2^2$.

However, if we define new variables $Z_1 = X_1$, $Z_2 = X_1^2$, $Z_3 = X_2$, and $Z_4 = X_2^2$, the boundary equation becomes

$$
Z_2 + Z_4 + 2Z_1 - 4Z_3 + 1 = 0,
$$

which is **linear in the transformed variables** $Z_1, Z_2, Z_3, Z_4$. Thus, by augmenting the feature space with quadratic terms, the originally nonlinear decision boundary becomes linear in the higher-dimensional space.

| Obs. | $X_1$ | $X_2$ | $Y$  |
| ---- | ----- | ----- | ---- |
| 1    | 3     | 4     | Red  |
| 2    | 2     | 2     | Red  |
| 3    | 4     | 4     | Red  |
| 4    | 1     | 4     | Red  |
| 5    | 2     | 1     | Blue |
| 6    | 4     | 3     | Blue |
| 7    | 4     | 1     | Blue |

**Sketch the observations.**

**(b) Sketch the optimal separating hyperplane, and provide the equation for this hyperplane (of the form (9.1)).**

**(d) On your sketch, indicate the margin for the maximal margin hyperplane.**

```python
import matplotlib.pyplot as plt

X1 = [3, 2, 4, 1, 2, 4, 4]
X2 = [4, 2, 4, 4, 1, 3, 1]
Y = ["Red", "Red", "Red", "Red", "Blue", "Blue", "Blue"]

plt.figure(figsize=(8, 6))

for x1, x2, y in zip(X1, X2, Y):
    color = "red" if y == "Red" else "blue"
    plt.scatter(x1, x2, c=color, s=100, edgecolors="k")

plt.axline(xy1=(2, 1.5), xy2=(4, 3.5), c='g', label=r"2X_1 - 2X_2 - 1 = 0")

plt.axline(xy1=(2, 2), xy2=(4, 4), c='c', ls='--', label="margin")
plt.axline(xy1=(2, 1), xy2=(4, 3), c='c', ls='--')

plt.scatter([2, 2, 4, 4], [1, 2, 3, 4], c='k', linewidths=5, label='Support Vectors')

plt.xlabel(r"$X_1$")
plt.ylabel(r"$X_2$")
plt.grid(True)
plt.legend()
plt.show()
```

**(c) Describe the classification rule for the maximal margin classifier. It should be something along the lines of “Classify to Red if $\beta_0 + \beta_1 X_1 + \beta_2 X_2 > 0$, and classify to Blue otherwise.” Provide the values for $\beta_0$, $\beta_1$, and $\beta_2$.**

The maximal margin classifier separates the two classes (Red and Blue) with a straight line that maximizes the margin between them. The decision boundary is defined by the equation:

$$
2X_1 - 2X_2 - 1 = 0.
$$

Following the general linear classifier form, we can write the rule as:

$$
\text{Classify to Red if } \beta_0 + \beta_1 X_1 + \beta_2 X_2 > 0, \text{ and classify to Blue otherwise.}
$$

Comparing with the boundary equation, we identify the coefficients:

$$
\beta_0 = -1, \quad \beta_1 = 2, \quad \beta_2 = -2.
$$

Thus, the classification rule is:

$$
\text{Classify to Red if } -1 + 2X_1 - 2X_2 > 0, \text{ otherwise classify to Blue.}
$$
