**(a) In a study of disease relapse, due to a careless research scientist, all patients whose phone numbers begin with the number “2” are lost to follow up.**

The censoring mechanism is independednt since the phone number of patients isn't related to the disease or any of the input variables.

**(b) In a study of longevity, a formatting error causes all patient ages that exceed 99 years to be lost (i.e. we know that those patients are more than 99 years old, but we do not know their exact ages).**

This is dependent censoring, because whether someone is censored depends on their survival time (longevity).

**(c) Hospital A conducts a study of longevity. However, very sick patients tend to be transferred to Hospital B, and are lost to follow up.**

This is dependent censoring because the censoring mechanism (transfer of very sick patients) is directly tied to the outcome (longevity).

**(d) In a study of unemployment duration, the people who find work earlier are less motivated to stay in touch with study investigators, and therefore are more likely to be lost to follow up.**

This is dependent censoring, because the probability of being censored depends on unemployment duration (the study’s outcome).

**(e) In a study of pregnancy duration, women who deliver their babies pre-term are more likely to do so away from their usual hospital, and thus are more likely to be censored, relative to women who deliver full-term babies.**

This is dependent censoring, because the chance of being censored is directly related to the outcome (pregnancy duration).

**(f) A researcher wishes to model the number of years of education of the residents of a small town. Residents who enroll in college out of town are more likely to be lost to follow up, and are also more likely to attend graduate school, relative to those who attend college in town.**

This is dependent censoring, because the probability of being censored is associated with the likely outcome (educational duration).

**(g) Researchers conduct a study of disease-free survival (i.e. time until disease relapse following treatment). Patients who have not relapsed within five years are considered to be cured, and thus their survival time is censored at five years.**

This is independent censoring, it is administrative censoring at a fixed follow-up time (study end) and does not depend on individual unobserved future failure times (assuming the 5-year cutoff is independent of subjects’ individual risks).

**(h) We wish to model the failure time for some electrical component. This component can be manufactured in Iowa or in Pittsburgh, with no difference in quality. The Iowa factory opened five years ago, and so components manufactured in Iowa are censored at five years. The Pittsburgh factory opened two years ago, so those components are censored at two years.**

This is independent censoring, because the censoring mechanism (factory opening time) is unrelated to the failure time of the component.

**(i) We wish to model the failure time of an electrical component made in two different factories, one of which opened before the other. We have reason to believe that the components manufactured in the factory that opened earlier are of higher quality.**

This is dependent censoring, because the censoring mechanism (factory opening time) is related to a variable (quality) that directly affects the outcome (failure time).

Number at risk just before each event:

**7. In this problem, we will derive (11.5) and (11.6), which are needed for the construction of the log-rank test statistic (11.8). Recall the notation in Table 11.1.**

**(a) Assume that there is no difference between the survival functions of the two groups. Then we can think of $q_{1k}$ as the number of failures if we draw $r_{1k}$ observations, without replacement, from a risk set of $r_k$ observations that contains a total of $q_k$ failures. Argue that $q_{1k}$ follows a hypergeometric distribution. Write the parameters of this distribution in terms of $r_{1k} , r_k,$ and $q_k$.**

We are comparing **survival times between two groups** (e.g., treatment vs. control) and want to test:

$$
H_0: \text{the survival distributions of the two groups are identical}.
$$

Suppose we have $K$ distinct observed event times $t_1, t_2, \dots, t_K$. At each event time $t_k$:

* $r_k$ = number of individuals at risk just before $t_k$ (both groups combined)
* $q_k$ = total number of events (failures) at $t_k$
* $r_{1k}$ = number of individuals in group 1 at risk at $t_k$
* $q_{1k}$ = number of events in group 1 at $t_k$

If the null hypothesis is true, **events occur randomly among all individuals at risk**, independent of group.

At each event time $t_k$, we can think of "drawing" $r_{1k}$ individuals from the $r_k$ total at-risk population **without replacement**, and counting how many of these $r_{1k}$ individuals experience the event.

* Total population at risk: $r_k$
* Total events among them: $q_k$
* Sample size: $r_{1k}$

This exactly matches the **hypergeometric distribution**, whose probability mass function is:

$$
P(X = x) = \frac{\binom{K}{x} \binom{N-K}{n-x}}{\binom{N}{n}}, \quad x = 0,1,2,\dots,n
$$

In our case:

$$
q_{1k} \sim \text{Hypergeometric}(N=r_k, K=q_k, n=r_{1k})
$$

where $N = r_k$ is the total risk set, $K = q_k$ is the total failures at $t_k$, and $n = r_{1k}$ is the group 1 risk set size.

**(b) Given your previous answer, and the properties of the distribution, what are the mean and variance of $q_{1k}$? Compare your answer to (11.5) and (11.6).**

**Property of hypergeometric distribution:** For $X \sim \text{Hypergeometric}(N, K, n)$,

$$
\mathbb{E}[X] = n \frac{K}{N}, \quad 
\text{Var}(X) = n \frac{K}{N} \frac{N-K}{N} \frac{N-n}{N-1}.
$$

Applying this to $q_{1k}$:

$$
\mathbb{E}[q_{1k}] = r_{1k} \frac{q_k}{r_k} \equiv \mu_k
$$

This is the expected number of events in group 1 at $t_k$ under $H_0$, which matches equation (11.5).

$$
\text{Var}(q_{1k}) = r_{1k} \frac{q_k}{r_k} \left(1 - \frac{q_k}{r_k}\right) \frac{r_k - r_{1k}}{r_k - 1} 
= \frac{q_k \, (r_{1k}/r_k) \, (1 - r_{1k}/r_k) \, (r_k - q_k)}{r_k - 1}
$$

This is exactly equation (11.6).

The **log-rank test** compares observed and expected events across all event times. Define:

$$
X = \sum_{k=1}^K q_{1k} \quad \text{(total observed events in group 1)}.
$$

* Expected total events under $H_0$:

$$
\mu = \sum_{k=1}^K \mathbb{E}[q_{1k}] = \sum_{k=1}^K r_{1k} \frac{q_k}{r_k}.
$$

* Approximate variance of the sum:

Even though the $q_{1k}$ are not independent, for large samples we approximate the variance as:

$$
\text{Var}(X) \approx \sum_{k=1}^K \text{Var}(q_{1k}) = \sum_{k=1}^K \frac{q_k (r_{1k}/r_k) (1 - r_{1k}/r_k) (r_k - q_k)}{r_k - 1}.
$$

Finally, the **log-rank test statistic** is standardized as:

$$
W = \frac{\sum_{k=1}^K (q_{1k} - \mu_k)}{\sqrt{\sum_{k=1}^K \text{Var}(q_{1k})}}.
$$

This is equation (11.8) in the text.

* $W$ measures the **difference between observed and expected events** in group 1, relative to the variance.
* Under $H_0$ (no survival difference), $W$ approximately follows a **standard normal distribution** for large samples.
* A two-sided p-value can be computed:

$$
p = 2 \, P(Z > |W|), \quad Z \sim N(0,1)
$$

* Large $|W|$ → strong evidence against $H_0$.

**8. Recall that the survival function $S(t)$, the hazard function $h(t)$, and the density function $f(t)$ are defined in (11.2), (11.9), and (11.11), respectively. Furthermore, define $F(t) = 1 − S(t)$. Show that the following relationships hold:**

$$
f(t) = dF(t)/dt
$$
$$
S(t) = exp\left(-\int_0^t h(u)du \right)
$$

By definition of the CDF:

$$
F(t) = 1 - S(t)
$$

Differentiating both sides with respect to $t$:

$$
\frac{d}{dt}F(t) = -\frac{d}{dt}S(t)
$$

But from the definition of $f(t)$:

$$
f(t) = -\frac{d}{dt}S(t)
$$

So we have:

$$
f(t) = \frac{d}{dt} F(t)
$$

From the definition of the hazard function:

$$
h(t) = \frac{f(t)}{S(t)}
$$

Rewriting:

$$
f(t) = h(t) S(t)
$$

But $f(t) = -\frac{d}{dt} S(t)$, so:

$$
-\frac{d}{dt} S(t) = h(t) S(t)
$$

Rewriting:

$$
\frac{dS(t)}{dt} = -h(t) S(t)
$$

This is a **first-order differential equation**. We can solve it by separating variables:

$$
\frac{dS(t)}{S(t)} = -h(t) \, dt
$$

Integrate both sides from 0 to $t$:

$$
\int_{S(0)}^{S(t)} \frac{dS}{S} = -\int_0^t h(u) \, du
$$

$$
\ln S(t) - \ln S(0) = - \int_0^t h(u) \, du
$$

Assuming $S(0) = 1$ (probability of surviving past time 0 is 1):

$$
\ln S(t) = - \int_0^t h(u) \, du
$$

Exponentiating both sides:

$$
S(t) = \exp\left( - \int_0^t h(u) \, du \right)
$$

**9. In this exercise, we will explore the consequences of assuming that the survival times follow an exponential distribution.**

**(a) Suppose that a survival time follows an $\text{Exp}(\lambda)$ distribution, so that its density function is $f(t) = \lambda \exp(-\lambda t)$. Using the relationships provided in Exercise 8, show that $S(t) = \exp(-\lambda t)$.**

We are given:

$$
f(t) = \lambda e^{-\lambda t}, \quad t \ge 0
$$

From Exercise 8, we know the relationship between the survival function $S(t)$ and the hazard function $h(t)$:

$$
S(t) = \exp\left(-\int_0^t h(u)\, du\right), \quad \text{with } h(t) = \frac{f(t)}{S(t)}
$$

We know $S(t) = 1 - F(t)$, but let’s solve directly: assume $S(t) = e^{-\lambda t}$ and check:

$$
h(t) = \frac{\lambda e^{-\lambda t}}{e^{-\lambda t}} = \lambda
$$

The hazard is constant, as expected for an exponential distribution.

$$
S(t) = \exp\Big(-\int_0^t \lambda du\Big) = \exp(-\lambda t)
$$

**(b) Now suppose that each of $n$ independent survival times follows an $\text{Exp}(\lambda)$ distribution. Write out an expression for the likelihood function (11.13).**

Let $y_1, \dots, y_n$ be observed survival times, and $\delta_i$ indicate censoring:

$$
\delta_i = 
\begin{cases} 
1 & \text{if the event is observed} \\
0 & \text{if censored} 
\end{cases}
$$

The likelihood for independent survival times is:

$$
L(\lambda) = \prod_{i=1}^n [f(y_i)]^{\delta_i} [S(y_i)]^{1-\delta_i}.
$$

For the exponential case:

$$
f(y_i) = \lambda e^{-\lambda y_i}, \quad S(y_i) = e^{-\lambda y_i}.
$$

So the likelihood becomes:

$$
L(\lambda) = \prod_{i=1}^n [\lambda e^{-\lambda y_i}]^{\delta_i} [e^{-\lambda y_i}]^{1-\delta_i} = \prod_{i=1}^n \lambda^{\delta_i} e^{-\lambda y_i}
$$

**(c) Show that the maximum likelihood estimator for $\lambda$ is**
$$
\hat{\lambda} = \frac{\sum_{i=1}^{n} \delta_i}{\sum_{i=1}^{n} y_i}.
$$

Take the log-likelihood:

$$
\ell(\lambda) = \sum_{i=1}^n \delta_i \ln \lambda - \lambda \sum_{i=1}^n y_i
$$

Differentiate with respect to $\lambda$ and set to 0:

$$
\frac{d\ell}{d\lambda} = \frac{\sum_{i=1}^n \delta_i}{\lambda} - \sum_{i=1}^n y_i = 0
$$

Solve for $\lambda$:

$$
\hat{\lambda} = \frac{\sum_{i=1}^n \delta_i}{\sum_{i=1}^n y_i}
$$

**(d) Use your answer to (c) to derive an estimator of the mean survival time.**

For an exponential distribution, the mean survival time is:

$$
\text{E}[T] = \frac{1}{\lambda}
$$

So the natural estimator is the reciprocal of $\hat{\lambda}$:

$$
\widehat{\text{E}}[T] = \frac{1}{\hat{\lambda}} = \frac{\sum_{i=1}^n y_i}{\sum_{i=1}^n \delta_i}
$$

This gives an intuitive result: the mean survival time is the total observed time divided by the number of events.
